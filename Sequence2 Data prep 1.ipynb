```python
df['year']= df['datetime'].dt.year
df['month']= df['datetime'].dt.month
df['day']= df['datetime'].dt.day
```

```python
dw_mapping={
    0: 'Monday', 
    1: 'Tuesday', 
    2: 'Wednesday', 
    3: 'Thursday', 
    4: 'Friday',
    5: 'Saturday', 
    6: 'Sunday'
} 
df['day_of_week_name']=df['datetime'].dt.weekday.map(dw_mapping)
```

```python
df.info()
```

```python
import numpy as np
df['postal_code'] =df['postal_code'].replace(np.nan, 0)
```

```python
df['postal_code'] = df['postal_code'].astype(int)
```

```python
df = df.set_index("outlet_type")
```

```python
df = df.drop('OUTOFORDER', axis=0)
df = df.drop('UNKNOWN', axis=0)
```

```python
df=df.reset_index()
```

```python
df.head()
```

```python
df = df.rename(columns={"outlet_type": "Status","day_of_week_name":"Week_Name"})
```

```python
df['Status'] = df['Status'].map({'AVAILABLE': 0, 'OCCUPIED': 1})
```

```python
df['Dates'] = pd.to_datetime(df['datetime']).dt.date
df['Dates']= pd.to_datetime(df['Dates'])
```

```python
df.to_csv('seq1.csv')
```
