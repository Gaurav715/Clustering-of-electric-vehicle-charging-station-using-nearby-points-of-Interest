```python
import pandas as pd
df =pd.read_pickle("seq1.pickle")
```

```python
df.head()
```

```python
df0=df[df.Status==0]
df1=df[df.Status==1]
```

```python
df
```

```python
df2= pd.merge(df1,df0,how='inner',left_on=['evse_id','Dates'],right_on=['evse_id','Dates'])
```

```python
df2
```

```python
df2.info()
```

```python
df3=df2[['evse_id','datetime_x','datetime_y','Status_x','Status_y','city_y']]
```

```python
df_= df.groupby(['evse_id','park_id']).mean().reset_index().drop(['year','month','day','Status'],axis=1)
```

```python
df5= pd.merge(df3,df_,how='inner',left_on=['evse_id'],right_on=['evse_id'])
```

```python
df5['datetime_y']=pd.to_datetime(df5['datetime_y'])
df5['datetime_x']=pd.to_datetime(df3['datetime_x'])
```

```python
df5['TD']=df5['datetime_y']-df5['datetime_x']
```

```python
s = df5['TD'].iloc[0]
```

```python
df5['Charging_Time_in_min']=df5['TD'].apply(lambda x: x.seconds/60)
```

```python
df5['Charging_Time_in_hours']=df5['Charging_Time_in_min']/60
```

```python
df5['postal_code'] = df5['postal_code'].astype('int')
```

```python
df5=df5.rename(columns = {'datetime_x':'datetime_start','datetime_y':'datetime_stop','Status_x':'Charging_Started'
                     ,'Status_y':'Charging_Stop','city_y':'city'})
```

```python
df5['year']= df5['datetime_start'].dt.year
df5['month']= df5['datetime_start'].dt.month
df5['day']= df5['datetime_start'].dt.day
```

```python
dw_mapping={
    0: 'Monday', 
    1: 'Tuesday', 
    2: 'Wednesday', 
    3: 'Thursday', 
    4: 'Friday',
    5: 'Saturday', 
    6: 'Sunday'
} 
df5['day_of_week_name_Start']=df5['datetime_start'].dt.weekday.map(dw_mapping)
```
